```python
import pyautogui
import time


#Passo a Passo do Projeto

#Comandos que serão utilizados:
# time.sleep(tempo em sec) --> pausa entre um código e outro 
# pyauotgui.PAUSE --> pausa para rodar cada código
# pyautogui.click --> clica com o mouse
# pyautogui.write --> escreve algo
# pyautogui.press --> aperta uma tecla
# pyautogui.hotkey --> aperta uma combinação de teclas. ex: crtl + s

pyautogui.PAUSE = 1  #1 segundo

#Passo 1: Com o navegador aberto, entrar no sistema da empresa(no link)

pyautogui.hotkey("ctrl", "t")
pyautogui.write("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")
pyautogui.press("enter")

#Passo 2: Fazer login

#clicar na área de login
pyautogui.click(x=917, y=351) # X e Y representam as coordenadas da área de login, para descobrir qual é a posição, usamos um "cronômetro" chamado pela função time.sleep(tempo em segundos) juntamente com a função pytautogui.position() que irá mostrar a posição exata da marcação. 
pyautogui.write("meu_login")

#clicar na área de inserir a senha
pyautogui.click(x=869, y=429)
pyautogui.write("minha_senha")

#clicar em acessar
pyautogui.click(x=918, y=495)

#Passo 3: Exportar a base de dados

#selecionar o arquivo csv do sistema e fazer o download
time.sleep(3) #esperar carregar a página ...
pyautogui.click(x=305, y=359)
pyautogui.click(x=1718, y=176)
pyautogui.click(x=1480, y=562)
time.sleep(3) #esperar fazer o download ...
```

```python
#Passo 4: Colocar os indicadores
import pandas as pd   #o "as" faz com que o pacote "pandas" seja renomeado.

tabela = pd.read_csv(r"C:\Users\MAFRA\Downloads\Compras.csv", delimiter=";") #o "r" no início é para que o Python leia o caminho exatamente  como foi escrito. # o sep organiza nossa tabela com a separação em ';' 
display(tabela) # o display é uma função que está apenas no Jupyter, seria como um "print" mas de uma forma mais organizada e esteticamente agradável.

total_gasto = tabela["ValorFinal"].sum()
quantidade = tabela["Quantidade"].sum()
preco_medio = total_gasto / quantidade

print(total_gasto)
print(quantidade)
print(preco_medio)
```
